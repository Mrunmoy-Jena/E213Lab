### Plot of angular distributions

```python
import matplotlib.pyplot as plt
%matplotlib inline
ctheta=np.linspace(-1,1,100)
diffcrs=1+(ctheta**2)
diffcrt=(1-ctheta)**(-2)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.ylim([0.9,2.1])
plt.xticks(np.arange(-1.0, 1.5, 0.5))
plt.xlabel(r"$\cos\theta$")
plt.ylabel(r"$\dfrac{d\sigma}{d\Omega}$")
plt.plot(ctheta,diffcrs,'b',label='s-channel')
plt.legend(loc='upper left')
plt.subplot(1,2,2)
plt.ylim([0,14])
plt.xticks(np.arange(-1.0, 1.5, 0.5))
plt.xlabel(r"$\cos\theta$")
plt.ylabel(r"$\dfrac{d\sigma}{d\Omega}$")
plt.plot(ctheta,diffcrt,'r-.',label='t-channel')
plt.plot(ctheta,diffcrs,'b',label='s-channel')
plt.plot(ctheta,diffcrs+diffcrt,'g--',label='total')
plt.legend(loc='upper left')
plt.tight_layout()
plt.savefig("angulardist.pdf", bbox_inches='tight')
plt.show()
```
